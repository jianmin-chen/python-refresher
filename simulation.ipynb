# Simulation

```python
import numpy as np
```

## Parameters

```python
density_water = 1000  # kg/m^3
density_object = 500  # kg/m^3
object_volume = 0.1  # m^3
g = 9.81  # m/s^2
time_step = 0.01  # s
time_max = 10  # s
```

```python
time = 0  # s
depth = 0  # m
velocity = 0  # m/s
acceleration = 0  # m/s^2
"""
    Buoyancy is an upward force exerted by a fluid that pushes an object upward?
    Weight (mass w/ gravity applied?) is going to be the
"""
buoyancy = (
    density_water * g * object_volume
)  # N: fb = (p = displacement) * (g = gravity) * (h = )
weight = density_object * object_volume * g  # N
```

```python
# lists for plotting later - fill w/ 0s based on time
time = np.arange(0, time_max, time_step)
position = np.zeros_like(time)
velocity = np.zeros_like(time)
acceleration = np.zeros_like(time)
```

```python
# simulation loop
for i in range(1, len(time)):
    # calculate the force on the object
    force = buoyancy - weight
    # calculate acceleration
    acceleration[i] = force / (
        weight / g
    )  # Acceleration is equal to F / m (if it were moving force would be affected?)
    # calculate velocity
    velocity[i] = acceleration[i - 1] * time_step + velocity[i - 1]
    # calculate position
    position[i] = velocity[i - 1] * time_step + position[i - 1]
```
